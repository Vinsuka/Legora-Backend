```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Legora-Evaluation/Embedding Comp/legal_doc_evaluation_results.csv")
df.head()
```

# F1 Score
The F1 Score is the harmonic mean of Precision and Recall.



```python
# Plot F1 Score across different Chunk Sizes
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Chunk Size", y="F1 Score", hue="Model", marker="o")
plt.title("F1 Score vs Chunk Size per Model")
plt.xlabel("Chunk Size")
plt.ylabel("F1 Score")
plt.grid(True)
plt.show()

# Average F1 Score per Model
avg_f1 = df.groupby('Model')['F1 Score'].mean().sort_values(ascending=False)
print("Average F1 Score per Model:")
print(avg_f1)
```

# Diversity
Diversity measures how different or varied the returned chunks are.

```python
# Plot Diversity across different Chunk Sizes
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Chunk Size", y="Diversity", hue="Model", marker="o")
plt.title("Diversity vs Chunk Size per Model")
plt.xlabel("Chunk Size")
plt.ylabel("Diversity Score")
plt.grid(True)
plt.show()

# Average Diversity per Model
avg_diversity = df.groupby('Model')['Diversity'].mean().sort_values(ascending=False)
print("Average Diversity Score per Model:")
print(avg_diversity)
```
